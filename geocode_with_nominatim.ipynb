```python
import pandas as pd
import requests
import time
```

```python
df = pd.read_excel("/content/Communes_cleaned_MADA.xlsx")
df.columns = df.columns.str.strip() # Clean column names
```

```python
df
```

```python
def get_country_code(pays):
    """Retourne le code ISO 3166-1 alpha-2 du pays"""
    country_codes = {
        "Madagascar": "mg"
    }
    return country_codes.get(pays, "")
```

```python
results = []

headers = {
    "User-Agent": "Baobab-geoloc/1.0 (afmbaye@baobab.com)"
}

for _, row in df.iterrows():
    commune = row["Commune"]
    pays = row["Pays"]

    # Geocodage sur commune + pays avec restriction au pays
    query = f"{commune}, {pays}".replace(" ", "+")
    url = f"https://nominatim.openstreetmap.org/search?q={query}&format=json&limit=1&countrycodes={get_country_code(pays)}&polygon_geojson=1&email=afmbaye@baobab.com"

    print(f"Trying: {commune}, {pays}")
    r = requests.get(url, headers=headers)

    try:
        data = r.json()
    except requests.exceptions.JSONDecodeError:
        print(f"JSONDecodeError for URL: {url}")
        print(f"Response text: {r.text}")
        data = []

    if data:
        lon = data[0].get("lon")
        lat = data[0].get("lat")
        place_id = data[0].get("place_id")
        osm_id = data[0].get("osm_id")
        osm_class = data[0].get("class")
        osm_type = data[0].get("type")
        addresstype = data[0].get("addresstype")
        name = data[0].get("name")
        display_name = data[0].get("display_name")
        boundingbox = data[0].get("boundingbox")
        status = "FOUND"
    else:
        lon = "NOT FOUND"
        lat = "NOT FOUND"
        place_id = "NOT FOUND"
        osm_id = "NOT FOUND"
        osm_class = "NOT FOUND"
        osm_type = "NOT FOUND"
        addresstype = "NOT FOUND"
        name = "NOT FOUND"
        display_name = "NOT FOUND"
        boundingbox = "NOT FOUND"
        status = "NOT FOUND"

    results.append({
        "commune": commune,
        "pays": pays,
        "lon": lon,
        "lat": lat,
        "place_id": place_id,
        "osm_id": osm_id,
        "class": osm_class,
        "type": osm_type,
        "addresstype": addresstype,
        "name": name,
        "display_name": display_name,
        "boundingbox": boundingbox,
        "status": status
    })

    time.sleep(1.1)

results_df = pd.DataFrame(results)
```

```python
results = []

headers = {
    "User-Agent": "Baobab-geoloc/1.0 (afmbaye@baobab.com)"
}

for _, row in df.iterrows():
  commune = row["Commune"]
  ville = row["Ville"]
  pays = row["Pays"]

  # Try commune + pays first
  query = (commune + ", " + pays).replace(" ", "+")
  url = f"https://nominatim.openstreetmap.org/search?q={query}&format=json&limit=1&polygon_geojson=1&email=your_actual_email@example.com"

  print(f"Trying: {query}")
  r = requests.get(url, headers=headers)

  try:
      data = r.json()
  except requests.exceptions.JSONDecodeError:
      print(f"JSONDecodeError for URL: {url}")
      print(f"Response text: {r.text}")
      data = []

  # If not found, try ville + pays
  if not data:
      query = (ville + ", " + pays).replace(" ", "+")
      url = f"https://nominatim.openstreetmap.org/search?q={query}&format=json&limit=1&polygon_geojson=1&email=your_actual_email@example.com"

      print(f"Fallback to: {query}")
      r = requests.get(url, headers=headers)

      try:
          data = r.json()
      except requests.exceptions.JSONDecodeError:
          print(f"JSONDecodeError for URL: {url}")
          data = []

  if data:
      lon = data[0].get("lon")
      lat = data[0].get("lat")
      place_id = data[0].get("place_id")
      osm_id = data[0].get("osm_id")
      osm_class = data[0].get("class")
      osm_type = data[0].get("type")
      addresstype = data[0].get("addresstype")
      name = data[0].get("name")
      display_name = data[0].get("display_name")
      boundingbox = data[0].get("boundingbox")
  else:
      lon = None
      lat = None
      place_id = None
      osm_id = None
      osm_class = None
      osm_type = None
      addresstype = None
      name = None
      display_name = None
      boundingbox = None

  results.append({
      "commune": commune,
      "ville": ville,
      "pays": pays,
      "lon": lon,
      "lat": lat,
      "place_id": place_id,
      "osm_id": osm_id,
      "class": osm_class,
      "type": osm_type,
      "addresstype": addresstype,
      "name": name,
      "display_name": display_name,
      "boundingbox": boundingbox
  })

  time.sleep(1.1)

results_df = pd.DataFrame(results)
```
